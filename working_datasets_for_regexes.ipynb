```python
class processBezrealitky():
    
    def __init__(self, dataset = data['bezrealitky']):
    
        self.df = dataset.dropna(subset = ['price'])
        self.df = self.df.drop(self.df.columns[[0]], axis=1)
        
        self.clean_dataset(self.df)
    
    # drop missing prices
    def get_meters(self, row):
        metraz = row.split(',')[-1].strip()
        metry_cislo = re.search(r'\d+', metraz)
        return int(metry_cislo.group(0)) if metry_cislo else ""

    def fix_price(self, row):
        cut_currency =  row.split(' ')[0]
        return int(cut_currency.replace('.', ''))

    #get meters as number
    def clean_dataset(self, df):
        df['area'] = df['area'].apply(self.get_meters)
    
    # price as number
    
        df['price'] = df['price'].apply(self.fix_price)
        
        return df
```

# PRAZSKE REALITY

```python
data['prazskereality'].head()
```

```python
class processPrazskereality():
    
    
    def __init__(self, dataset = data['prazskereality']):
    
        self.df = dataset.dropna(subset = ['price'])
        self.df = self.df.drop(self.df.columns[[0]], axis=1)
        self.df = self.df.drop(self.df[self.df['price'] == 'Info o ceně u RK'].index)
        
        self.clean_dataset(self.df)

    def fix_price(self, row):
        cut_currency =  row[:-3]
        return cut_currency.replace('\xa0', '')

    def get_meters(self, row):
        metraz = row.split(',')[-1].strip()
        metry_cislo = re.search(r'\d+', metraz)
        return int(metry_cislo.group(0)) if metry_cislo else ""
    
    def get_region(self, row):
        if 'Praha' in str(row):
            region_street = row.split(' - ')[1]
            region = region_street.split(',')[0]
        return region
    
    def get_city(self, row):
        if 'Praha' in str(row):
            city = 'Praha'
            return city
        
    def get_street(self, row):
        street = row.split(',')[-1].strip()
        return street
    
    #get meters as number
    def clean_dataset(self, df):
        df['area'] = df['area'].apply(self.get_meters)
        df['price'] = df['price'].apply(self.fix_price)
        df['city_part'] = df['address'].apply(self.get_region)
        df['city'] = df['address'].apply(self.get_city)
        df['street'] = df['address'].apply(self.get_street)

        
        return df
```

```python
data['bytysReality'].head()
```

```python
data['bytysReality'] = data['bytysReality'].rename(columns = {'lokace': 'address', 
                                                              'cena': 'price',
                                                              'uzitna_plocha': 'area',
                                                              'podlazi': 'floor',
                                                              'url': 'link',
                                                              'parkovani': 'parking',
                                                              'popis': 'description',
                                                              'podlazi': 'floor',
                                                              'vytah': 'elevator',
                                                              'lodzie': 'loggia'})
```

```python
class processSreality():
    
    def __init__(self, dataset = data['bytysReality']):
    
        self.df = dataset.dropna(subset = ['price'])
        self.df = self.df.drop(self.df[self.df['price'] == 'Info o ceně u RK'].index)

        
        self.clean_dataset(self.df)

    def fix_price(self, row):
        cut_currency =  row[:-3]
        return cut_currency.replace('\xa0', '')
    
    def get_meters(self, row):
        metry_cislo = re.search(r'^[0-9]+', row)
        return int(metry_cislo.group(0)) if metry_cislo else ""
    
    def get_region(self, row):
        if 'Praha' in str(row):
            kraj = row.split(' - ')[-1]
        return kraj
    
    def get_city(self, row):
        if 'Praha' in str(row):
            city = 'Praha'
            return city
        
    def get_street(self, row):
        street = row.split(',')[0]
        return street
    
    def get_size(self, row):
        size = re.search(r'[\d][\+][\w]+', row)[0]
        return size
        
    #get meters as number
    def clean_dataset(self, df):
        df['area'] = df['area'].apply(self.get_meters)
        df['city_part'] = df['address'].apply(self.get_region)
        df['price'] = df['price'].apply(self.fix_price)
        df['city'] = df['address'].apply(self.get_city)
        df['street'] = df['address'].apply(self.get_street)
        df['size'] = df['title'].apply(self.get_size)
        
        return df
```

# FLOOR REGEX


```python
num_wordings = ['první', 'druhé', 'třetí', 'čtvrté', 'páté', 'šesté', 'sedmé', 'osmé', 'deváté', 'desáté',
               'jedenácté', 'dvanácté', 'třinácté']

mapping = {'první': 1, 'druhé': 2, 'třetí': 3, 'čtvrté': 4, 'páté': 5, 'šesté': 6, 'sedmé': 7, 'osmé': 8, 
           'deváté': 9, 'desáté': 10, 'jedenácté': 11, 'dvanácté': 12, 'třinácté': 13}

def get_floor(row):
    patro = re.search(r'([\w]+)(\s)([\w.]+)(\s)(pat[řero]+)', row)
    if patro is not None:
        patro = patro[0]
        number = re.search(r'\d', patro)
        if number is not None:
            patro = number[0]
        else:
            for word in num_wordings:
                if word in patro.strip():
                    patro = mapping[word]
                    break
        if 'poslední' in str(patro).strip():
            patro = 'last'
        if 'přízem' in str(patro).strip():
            patro = 'ground'
    return patro
```

```python
df['floor_desc'] = df['description'].apply(get_floor)
```
